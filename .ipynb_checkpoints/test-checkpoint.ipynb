```python
import pandas as pd
import tensorflow as tf

df = pd.read_csv("./archive-3/Datasets/2000-01.csv")
print(df.head())
```

```python
teams = set(df["HomeTeam"].unique())
dict_teams = {}
for team in teams:
    dict_teams[team] = [0, 0, 0]
for index, row in df.iterrows():
    dict_teams[row["HomeTeam"]][1] += row["FTHG"]
    dict_teams[row["HomeTeam"]][2] += row["FTAG"]
    dict_teams[row["AwayTeam"]][1] += row["FTAG"]
    dict_teams[row["AwayTeam"]][2] += row["FTHG"]
    if row["FTHG"] > row["FTAG"]:
        dict_teams[row["HomeTeam"]][0] += 3
    elif row["FTHG"] == row["FTAG"]:
        dict_teams[row["HomeTeam"]][0] += 1
        dict_teams[row["AwayTeam"]][0] += 1
    else:
        dict_teams[row["AwayTeam"]][0] += 3
print(dict_teams)
```
